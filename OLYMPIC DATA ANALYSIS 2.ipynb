```python
merged2=pd.concat([merged,pd.get_dummies(merged['Medal'])],axis=1)
merged2
```

```python
each_medal_df=merged2.groupby('region')[['Gold','Silver','Bronze']].sum()
total_medal_df=merged2.dropna(subset=['Medal']).groupby('region')['Medal'].count().reset_index()
number_medals_df=each_medal_df.merge(total_medal_df,on='region')
top10countries_medal_df=number_medals_df.set_index('region').sort_values('Medal',ascending=False).head(10)
#print(each_medal_df,total_medal_df,number_medals_df,top10countries_medal_df)
```

```python
top10countries_medal_df
```

```python
total_athletes=print("{} countries have participated in olympics.".format(merged2.region.nunique()))
total_athletes
```

```python
top10countries_medal_df[['Gold','Silver','Bronze']].plot(kind='bar', stacked=True,title='Top 10 countries got the most medals', xlabel='Country');
```

```python
athletes_by_year=merged2.drop_duplicates(['Year','Name'])['Year'].value_counts().reset_index().sort_values('index')
athletes_by_year=athletes_by_year.rename(columns={'index':'Year','Year':'total_athletes'})
```

```python
plt.plot(athletes_by_year.Year,athletes_by_year.total_athletes, 'b-o')
plt.title('Number of athletes participated')
```

```python
countries_by_year=merged2.drop_duplicates(['Year','region'])['Year'].value_counts().reset_index().sort_values('index')
countries_by_year=countries_by_year.rename(columns={'index':'Year','Year':'total_countries'})
plt.plot(countries_by_year.Year,countries_by_year.total_countries, 'r-o')
plt.title('Number of countries participated')
```

```python
top_medalists=merged2.dropna(subset=['Medal']).groupby('Name')['Medal'].count().reset_index().sort_values('Medal',ascending=False).head(20)
add_df=merged2.drop_duplicates('Name')[['Name','region','Sport']]
top_medalists_df=top_medalists.merge(add_df,on='Name',how='left')

top_medalists_df
```
